```python
files_test = glob.glob("selfies/*.jpg", recursive=True)
```

```python
docs_test = glob.glob("documents/*.jpg", recursive=True)
```

```python
SIZE = 64

raw_img = tf.io.read_file(f"{files_test[0]}")
img = tf.io.decode_image(raw_img, dtype=tf.float32, channels=3)
img_gray = tf.image.rgb_to_grayscale(img)
image_resized = tf.image.resize(img_gray, (SIZE, SIZE))
train_images = np.asarray(image_resized).reshape(1, SIZE, SIZE, 1)
train_images_list = []

#train_images_list.append(train_images

j = 1
for  i in (files_test[1:] + docs_test):
  raw_img = tf.io.read_file(f"{i}")
  img = tf.io.decode_image(raw_img, dtype=tf.float32, channels=3)
  img_gray = tf.image.rgb_to_grayscale(img)
  image_resized = tf.image.resize(img_gray, (SIZE, SIZE))
  train_sample = np.asarray(image_resized).reshape(1, SIZE, SIZE, 1)
  #train_images = np.concatenate((train_images, train_sample))
  train_images_list.append(train_sample)

  j += 1
  if (j % 1000) == 0:
     
    print(j)
```

```python
#np.array(train_images_list).shape
```

```python
joblib.dump(train_images_list, "train_images_list")
```

```python
for i in train_images_list:
  train_images = np.concatenate((train_images, i))
```

```python
 train_images.shape
```

```python
train_images

plt.imshow(np.asarray(np.squeeze(train_images[19451])), 'gray' )
```

```python
train_images.shape
```

```python
label_selfies = [0 for i in range(len(files_test))] 
docs_selfies = [1 for i in range(len(docs_test))] 
```

```python
train_labels = np.array(label_selfies + docs_selfies).reshape(-1, 1)
```

```python
np.save(open('train_images.npy', 'wb'), train_images)
np.save(open('train_labels.npy', 'wb'), train_labels)
```

```python
train_labels
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(SIZE, SIZE, 1)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Conv2D(64, (3, 3), activation='relu'))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(2))


model.compile(optimizer='adam',
              #loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              loss='mse',
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10 )
```
